```python
import pandas as pd
```

```python
cars={
    "car_brand":['Mecedece',"Honda","Honda","Toyota","Mecedece","Oudi"],
    "price":[1000000,200000,150000,250000,1000000,3000000],
    "safety":[8,6,7,9,5,10]
  }
cars=pd.DataFrame(cars)
cars
```

```python
cars.sort_index()
```

```python
cars.groupby(['car_brand']).count()
```

```python
cars.groupby(['car_brand'])['price'].agg(['min','max'])
```

```python
cars.groupby(['car_brand'])['price'].agg(['min','max']).sort_values(by='car_brand')
```
